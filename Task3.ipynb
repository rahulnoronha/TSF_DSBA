## **Step 3: <ins>Data Cleaning</ins>**

We need to check if the data has any missing values, and if they do we need to deal with these missing values. If not we can proceed with the next step.

```python
null_values = data_frame.isnull().sum()#isnull detect missing values and sum is used to find the total number of these missing values.
print(null_values)#Print the number of null values in the DataFrame.
```

```python
#If the DataFrame has no Null values then we do not need to do any processing of Null values
full_series = null_values[null_values==0]
print(len(full_series))
if (len(full_series)==len(null_values)):
    pass
else:
  data_frame = data_frame.dropna(axis=0, how='any')#Drop any row (axis=0) which has one or more Null values (how= 'any').
  print(data_frame.to_markdown())#Print the data which has been formatted to not include anymore Null values and proceed to the further steps.
```

Since the number of values in full series (series with 0 null values and the number of values in the data frame are equal this means we have no null values in the table and we can proceed.

## **Step 4: <ins>Analyse the Relationship between Variables </ins>**

```python
#Pairwise plot using Seaborn library
sns.pairplot(data_frame)
plt.show()
```

### Which shipping mode is more widely used?

```python
#We need to plot countplot of the shipping mode.
sns.countplot(x=data_frame['Ship Mode'],order=data_frame['Ship Mode'].value_counts().index)
plt.show()
```

```python
#We need to plot the pie chart of the shipping modes used.
data_frame['Ship Mode'].value_counts().plot.pie(autopct='%0.2f%%',radius=2)
plt.show()
```

We notice that Standard class is the most widely used Ship Mode, followed by Second Class.

### Which Segment products are sold more?

```python
#We need to plot countplot of the Segment.
sns.countplot(x=data_frame['Segment'],order=data_frame['Segment'].value_counts().index)
plt.show()
```

```python
#We need to plot the pie chart of the Segment.
data_frame['Segment'].value_counts().plot.pie(autopct='%0.2f%%',radius=2)
plt.show()
```

We can observe that the Consumer segment products are the most sold followed by Corporate segment products.

### **Which are the most sold Categories and the least sold Categories**

```python
#Create a pie chart having the division of categories sold
data_frame['Category'].value_counts().plot.pie(autopct='%0.2f%%', radius=2)
plt.show()
```

```python
#Showing a countplot of the categories
sns.countplot(x=data_frame['Category'],order=data_frame['Category'].value_counts().index)
plt.show()
```

We notice that the most sold category is the Office supplies category and the least sold category is the Technology category.

### **What is the most sold Sub-Category?**

```python
#Create a pie chart having the division of sub-categories sold
data_frame['Sub-Category'].value_counts().plot.pie(autopct='%0.2f%%', radius=2)
plt.show()
```

```python
#Showing a countplot of the categoriest.
plt.figure(figsize=(20,5))
sns.countplot(x=data_frame['Sub-Category'], order=data_frame['Sub-Category'].value_counts().index)
plt.show()
```

Here we notice that the most sold sub-category of products are Binders.

### **Which State, City and Region have the most and least Orders?**

```python
### Plot the orders in the state in a pie chart
plt.figure(figsize=(20,5))
sns.countplot(x=data_frame['State'], order=data_frame['State'].value_counts().head(15).index)#Observe first 15 values.
plt.show()
plt.figure(figsize=(20,5))
sns.countplot(x=data_frame['State'], order=data_frame['State'].value_counts().tail(15).index)#Observe last 15 values
plt.show()             
```

```python
### Plot the orders in the city in a pie chart
plt.figure(figsize=(20,5))
sns.countplot(x=data_frame['City'], order=data_frame['City'].value_counts().head(15).index)#Observe first 15 values.
plt.show()
plt.figure(figsize=(20,5))
sns.countplot(x=data_frame['City'], order=data_frame['City'].value_counts().tail(15).index)#Observe last 15 values
plt.show()             
```

```python
### Plot the orders in the region in a pie chart
plt.figure(figsize=(20,5))
sns.countplot(x=data_frame['Region'], order=data_frame['Region'].value_counts().index)#Observe first 15 values.
plt.show()         
```

We notice
1. **The state with the most orders is California and the one with the least number of orders is Wyoming.**
1. **The city with most orders in New York City and there are many states with are present with the least number of orders.**
1. **The region with the most orders is the West region and the region with the least orders is South.**

### **Bar Graph of the Average Profit or Loss and the Sales of a Category.**

```python
data_frame.groupby("Category")[[ 'Sales','Profit']].sum().plot.bar(color = ["blue","yellow"],figsize= (10,5))
plt.ylabel("Profit /loss and Sales")

plt.show()
```

We notice the most profit is made in the Technology category.

### **Bar Graph of the Average Profit or Loss and the Sales of a City.**

```python
#Top 50 by Sales
profit_loss_city=data_frame.groupby('City').agg({'Sales':'sum','Profit':'sum'}).sort_values(by = "Sales", ascending = False).head(50)
profit_loss_city.plot.bar(color = ["green", "yellow"],figsize= (18,15))
plt.title("Profit/Loss & Sales in bottom 50 City")
plt.xlabel("City")
plt.ylabel("Profit/Loss & Sales")
plt.show()
#Bottom 50 by Sales
profit_loss_city=data_frame.groupby('City').agg({'Sales':'sum','Profit':'sum'}).sort_values(by = "Sales", ascending = False).tail(50)
profit_loss_city.plot.bar(color = ["red", "blue"],figsize= (18,15))
plt.title("Profit/Loss & Sales in bottom 50 City")
plt.xlabel("City")
plt.ylabel("Profit/Loss & Sales")
plt.show()
```

```python
#Top 50 by Sales
profit_loss_State=data_frame.groupby('State').agg({'Sales':'sum','Profit':'sum'}).sort_values(by = "Sales", ascending = False).head(50)
profit_loss_State.plot.bar(color = ["green", "yellow"],figsize= (18,15))
plt.title("Profit/Loss & Sales in top 50 State")
plt.xlabel("State")
plt.ylabel("Profit/Loss & Sales")
plt.show()
#Bottom 50 by Sales
profit_loss_State=data_frame.groupby('State').agg({'Sales':'sum','Profit':'sum'}).sort_values(by = "Sales", ascending = False).tail(50)
profit_loss_State.plot.bar(color = ["red", "blue"],figsize= (18,15))
plt.title("Profit/Loss & Sales in bottom 50 State")
plt.xlabel("State")
plt.ylabel("Profit/Loss & Sales")
plt.show()
```
